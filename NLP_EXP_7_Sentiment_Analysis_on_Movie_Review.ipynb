```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk import pos_tag

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('vader_lexicon')
```

```python
def clean_text(text):
    text = text.lower()
    text = ''.join([char for char in text if char not in [".",",","?"]])
    stop_words = stopwords.words('english')
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

def stem_text(text):
    porter = PorterStemmer()
    return ' '.join([porter.stem(word) for word in text.split()])

def get_pos_tags(text):
    return pos_tag(text.split())
```

```python
analyzer = SentimentIntensityAnalyzer()

for review, label in zip(data["Review"], data["Sentiment"]):
    clean_review = clean_text(review)
    stemmed_review = stem_text(clean_review)
    pos_tags = get_pos_tags(stemmed_review)
    
    scores = analyzer.polarity_scores(stemmed_review)
    
    compound_score = scores['compound']
    if compound_score >= 0.05:
        sentiment = "positive"
    else:
        sentiment = "negative"
    
    print(f"Review: {review[:500]}...")
    print(f"Label: {label}")
    print(f"Predicted Sentiment: {sentiment}")
    print(f"Compound Score: {compound_score}\n")
```
