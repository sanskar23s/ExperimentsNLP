```python
import tweepy
from textblob import TextBlob

# Open the file containing the tweets and values
with open('Exp_8_Twitter_Data.csv', 'r') as f:
    for line in f:
        try:
          # Extract the tweet text and sentiment value
          tweet_text, sentiment_value = line.strip().split(',')

          # Perform sentiment analysis using TextBlob
          text_blob = TextBlob(tweet_text)
          sentiment = text_blob.sentiment.polarity

        # Compare the sentiment value from the file with the sentiment value from TextBlob
          if int(sentiment_value) == 1:
              if sentiment > 0:
                  print("Correctly identified positive sentiment")
              else:
                  print("Incorrectly identified negative sentiment")
          elif int(sentiment_value) == 0:
              if abs(sentiment) < 0.01:
                  print("Correctly identified neutral sentiment")
              else:
                  print("Incorrectly identified non-neutral sentiment")
          elif int(sentiment_value) == -1:
              if sentiment < 0:
                  print("Correctly identified negative sentiment")
              else:
                  print("Incorrectly identified positive sentiment")
        except ValueError as e:
          continue
```
